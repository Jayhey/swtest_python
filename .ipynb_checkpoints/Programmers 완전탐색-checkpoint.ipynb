```python
from itertools import permutations

numbers = str(input())

num = [str(d) for d in numbers]


def is_prime(n):
    if n == 0 or n == 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    else:
        return True
per_total = []
for i in range(1,len(num)+1):
    per = list(map(int, list(map(''.join, list(itertools.permutations(num, i))))))
    per_total.extend(list(set(per)))
per_total = list(set(per_total))

tmp = []
for n in per_total:
    tmp.append(is_prime(n))
sum(tmp)
```

```python
from itertools import product

num = [[123, 1, 1], [356, 1, 0], [327, 2, 0], [489, 0, 1]]
cand = []
for a in product(['1','2','3','4','5','6','7','8','9'],repeat=3):
    cand.append(list(a))
cand = list(map(''.join, cand))
```

```python
def same_score(target, stand):
    target = [int(a) for a in target]
    #stand = [int(a) for a in stand]
    strike = 0
    ball = 0
    for i in range(len(stand)):
        for j in range(len(target)):
            if stand[i] == target[i]:
                strike += 1
                break
            elif i != j and stand[i] == target[j]:
                ball +=1
    return [strike, ball]
            
        
```

```python
for i in range(len(num)):
    newcand = []
    stand = [int(d) for d in str(num[i][0])]
    for c in cand:
        if same_score(c ,stand) == [num[i][1], num[i][2]]:
            newcand.append(c)
    cand = newcand[:]

print(cand)
    

```

```python

#4 123 1 1 356 1 0 327 2 0 489 0 1
a = [4,123,1,1,356,1,0,327,2,0,489,0,1]
count = 1
candid = []
for i in range(1,1000):
    if i < 10:
        candid.append('0'+'0'+str(i))
    elif 10<=i<100:
        candid.append('0'+str(i))
    else:
        candid.append(str(i))

#matching function

def matching(target,str_b):
    tar = [int(aa) for aa in target]
    b = [int(bb) for bb in str_b]
    strike = 0
    ball = 0
    for i in range(len(b)):
        for j in range(len(target)):
            if b[i] == tar[i]:
                strike += 1
                break
            elif i!=j and b[i]==tar[j]:
                ball += 1
    return [strike,ball]


for iter in range(a[0]):
    newcandid = []
    standard = 3*iter+1
    stand = [int(i) for i in str(a[standard])]
    for i in candid:
        if matching(i,stand) == [a[3*iter+2],a[3*(iter+1)]]:
            newcandid.append(i)
    candid = newcandid[:]

print(candid)

```
